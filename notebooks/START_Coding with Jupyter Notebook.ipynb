<h2>What is Jupyter Notebook?

Jupyter Notebook is an open-source web app provided through the [Project Jupyter](https://jupyter.org/) initiative. It can be used to create and share documents known as 'notebooks'. These notebooks can contain live code, data visualisations, images, text, equations and more. They have several use cases, including publishing work, sharing code, working collaborative and teaching.

<h4><i>Inserting cells</h4></i>

A new cell can be created by selecting the `+` button at the top toolbar, which inserts a new cell directly below the current cell you are working on. Alternatively, you can use the Insert dropdown menu to insert a cell either above or below the current cell.

<h4><i>Deleting cells</h4></i>

A cell can be deleted by selecting the `scissors` button at the top toolbar. To delete multiple cells, hold the `SHIFT` key with the `up` or `down` arrows (depending on whether you want to select cells above/below the current cell) and select the `scissors` button.

<h4><i>Edit vs. command mode<br></h4></i>
A selected cell with a <span style="color:green">green</span> border is in 'edit' mode, meaning that you can add or edit existing code/text. A selected cell with a <span style="color:blue">blue</span> border is in 'command' mode, meaning that you can use them with keyboard shortcuts (covered below).

You can toggle between the edit and command modes using the `ESCAPE` and `ENTER` keys. `ESCAPE` will switch to command mode; `ENTER` switches to edit mode.

<h2>Keyboard shortcuts</h2><br>
Keyboard shortcuts allow us to perform common, simple tasks with the keyboard.

Most shortcuts can only be used when the active cell is in <b>command mode</b>, so you may need to switch to command mode using the `ESCAPE` key (remember that a  <span style="color:blue">blue</span> border is used for command mode, whereas a  <span style="color:green">green</span> border signifies edit mode). Some important shortcuts you can use in command mode include:
* `SHIFT` and `ENTER`: runs the cell
* `ESCAPE` or `ENTER`: toggles between command/edit modes
* `UP` or `DOWN`: scrolls up/down to the next cell
* `SHIFT` + `UP` or `DOWN`: selects multiple cells above/below the active cell
* `SHIFT` + `m`: merges cells into one cell (if multiple cells are selected)
* `a` or `b`: inserts a cell above (a) or below (b) active cell
* `m` or `y`: switches between markdown cell (m) and code cell (y)
* `d`, `d`: deletes active cell (and use `z` to undo the delete)
* `h`: brings up list of keyboard shortcuts

There are few shortcuts when working in <b>edit mode</b>, however a particularly useful is: <br>
`CTRL`, `SHIFT` + `-`: splits the active cell where the cursor is currently placed.

<h4><i>Equations</h4></i>

Equations can be written in LaTeX style by wrapping the phrase with a `$` either side. Exponents need to have a `^` before the power to make them appear above the number. For example:

<h4><i>Escaping 'special characters'</h4></i>

Sometimes you don't special characters (e.g. *, [], ()) to be treated as special characters but to actually be displayed in your text. You can 'escape' special characters (i.e. treat them as normal characters) by putting a backslash (`\`) before you type the character in.

For example, if you wanted to wrap asterisks around a word <b>without</b> making it italic, you could add a backslash before each asterisk: This word should have \*asterisks\* around it (no italics!).

<h2>Kernels</h2>

A kernel is a program that runs and inspects your code. For every document you open, Jupyter Notebook starts a kernel that runs the code for that document. Each kernel can run code in a single programming language. By default Jupyter Notebook uses a Python kernel (ipykernel), but there are kernels for other languages (e.g. R, Julia and Ruby).

It's important to note that, once code has been executed within the kernel, the output exists across 'time and space' - meaning that if you run an individual line of code, the variables that you create in there will exist anywhere else in the document. What matters in Jupyter Notebook is therefore not the order of the *cells*, but the order in which you *run* the cells.

You may at times need to restart your kernel (and therefore rerun your program). Use the Kernel dropdown menu and select any of the following:
* Restart - restarts program
* Restart & clear output - restarts program and clears any output currently in the notebook
* Restart & run all - restarts program and runs the whole notebook (order is important here -- the program runs cell-by-cell from top-to-bottom)

*Example: accessing a variable before running it*

```python
print(myvariable)
```

```python
myvariable = "hello"
```

<h2>Error handling</h2>

Go to the Kernel dropdown menu and select 'restart and run all'. There should be an error under the code cell: print(myvariable). 

Jupyter Notebook displays errors as output under the corresponding code cell. It will also interrupt (stop) your program from running any further (unless you introduce error handling into your code). Notice how the margin for the code cell below (myvariable = "hello") doesn't have a number in the square brackets; this is because that cell was *not run*. The program stopped when it found an error at print(myvariable) and did not run any cells beyond this. 

<h2>Saving, sharing and exporting</h2>

To save notebooks, you can either use `CTRL` + `s` keys or File --> 'Save and Checkpoint'. This will save your notebook and create a 'checkpoint', which is a simple versioning control tool provided by Jupyter Notebook. Checkpoint versions of your file will be saved in the 'checkpoints' folder that Jupyter creates by default.

Sharing notebooks:
* Use File --> 'Download as...' to download the notebook in your chosen format (most often .ipynb or .py file). Note that some download options require installation of the nbconvert tool (e.g. downloading as PDF or HTML).
* Git can be used for version control. Combined with GitHub, you can incorporate version control and share your notebooks online.
* Google Colab allows you to upload .ipynb files, which can be easily shared with others. The advantage of Google Colab is that your audience do not need to install Python (or any imported libraries) in order to run your program.
